```python
import spacy
from spacy.matcher import Matcher
import os
import pdfplumber

# Load the English language model
nlp = spacy.load("en_core_web_sm")

# Define patterns for names, emails, phone numbers, and skills
name_pattern = [{"POS": "PROPN"}, {"POS": "PROPN"}]
email_pattern = [{"LIKE_EMAIL": True}]
phone_pattern = [{"ORTH": "("}, {"SHAPE": "ddd"}, {"ORTH": ")"}, {"SHAPE": "ddd"}, {"ORTH": "-", "OP": "?"}, {"SHAPE": "dddd"}]
skills_pattern = [{"LOWER": {"IN": ["python", "javascript", "react", "sql", "java", "html", "css", "git", "docker","cpp","Machine Learning","Power BI"]}}]

# Initialize the Matcher
matcher = Matcher(nlp.vocab)
matcher.add("NAME", [name_pattern])
matcher.add("EMAIL", [email_pattern])
matcher.add("PHONE", [phone_pattern])
matcher.add("SKILLS", [skills_pattern])

# Define the folder paths for resumes in PDF format and the job description
resume_folder_path = '/content/Resume/'
job_description_path = '/content/MAQ Software_Job Description_Associate Software Engineer (1).pdf'

# Function to extract text from a PDF file using pdfplumber
def extract_text_from_pdf(pdf_file_path):
    with pdfplumber.open(pdf_file_path) as pdf:
        text = ""
        for page in pdf.pages:
            text += page.extract_text()
    return text

# Process and match skills in each resume PDF file
for resume_filename in os.listdir(resume_folder_path):
    if resume_filename.endswith('.pdf'):
        resume_path = os.path.join(resume_folder_path, resume_filename)

        resume_text = extract_text_from_pdf(resume_path)

        if resume_text:
            doc = nlp(resume_text)
            matches = matcher(doc)

            print(f"Skills in {resume_filename}:")
            for match_id, start, end in matches:
                if nlp.vocab.strings[match_id] == "SKILLS":
                    skill = doc[start:end].text
                    print(f"- {skill}")

    print()

# Process and match skills in the job description file (PDF)
job_description_text = extract_text_from_pdf(job_description_path)

if job_description_text:
    doc = nlp(job_description_text)
    matches = matcher(doc)

    print("Skills in Job Description:")
    for match_id, start, end in matches:
        if nlp.vocab.strings[match_id] == "SKILLS":
            skill = doc[start:end].text
            print(f"- {skill}")
```

```python
import os
from PyPDF2 import PdfReader
import pdfplumber
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# First, install the necessary libraries if not already installed
!pip install PyPDF2 pdfplumber

# Define a function to calculate similarity between two texts
def calculate_similarity(text1, text2):
    content = [text1, text2]
    cv = CountVectorizer()
    matrix = cv.fit_transform(content)
    similarity_matrix = cosine_similarity(matrix)
    similarity_score = similarity_matrix[0][1] * 100
    return similarity_score

# Define the file path for the job description
jd_file_path = '/content/MAQ Software_Job Description_Associate Software Engineer (1).pdf'

# Define the folder path containing all the resumes
resume_folder_path = '/content/Resume/'

# Initialize a dictionary to store similarity scores for each resume
similarity_scores = {}

# Extract text from the job description
try:
    with pdfplumber.open(jd_file_path) as pdf:
        pages = pdf.pages[0]
        job_description = pages.extract_text()
except:
    job_description = ""

# Iterate through all resume files in the folder
for resume_filename in os.listdir(resume_folder_path):
    if resume_filename.endswith('.pdf'):
        resume_path = os.path.join(resume_folder_path, resume_filename)

        try:
            with pdfplumber.open(resume_path) as pdf:
                pages = pdf.pages[0]
                resume_text = pages.extract_text()
        except:
            resume_text = ""

        # Calculate similarity between the resume and the job description
        similarity = calculate_similarity(job_description, resume_text)

        # Store the similarity score in the dictionary
        key = f"Resume {resume_filename}"
        similarity_scores[key] = similarity

# Print the similarity scores for all resumes
for key, score in similarity_scores.items():
    print(f"{key}: Similarity Score with Job Description = {score:.2f}%")
```

```python
# Import required libraries
import os
from PyPDF2 import PdfReader
import re
import string
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

# Function to extract text from a PDF file
def extract_text_from_pdf(pdf_file_path):
    pdf_reader = PdfReader(pdf_file_path)
    text = ""
    for page in pdf_reader.pages:
        text += page.extract_text()
    return text

# Function to clean and analyze text
def analyze_resume_text(text, terms):
    # Convert text to lowercase
    text = text.lower()

    # Remove numbers
    text = re.sub(r'\d+', '', text)

    # Remove punctuation
    text = text.translate(str.maketrans('', '', string.punctuation))

    # Initialize score counters for each area
    quality = 0
    operations = 0
    supplychain = 0
    project = 0
    data = 0
    healthcare = 0

    # Calculate scores for each area
    for area in terms.keys():
        for word in terms[area]:
            if word in text:
                if area == 'Quality/Six Sigma':
                    quality += 1
                elif area == 'Operations management':
                    operations += 1
                elif area == 'Supply chain':
                    supplychain += 1
                elif area == 'Project management':
                    project += 1
                elif area == 'Data analytics':
                    data += 1
                else:
                    healthcare += 1

    return [quality, operations, supplychain, project, data, healthcare]

# Specify the folder containing resumes (PDF files)
resume_folder = '/content/Resume/'

# Create an empty list to store scores for each resume
resume_scores = []

# Create dictionary with industrial and system engineering key terms by area
terms = {'Quality/Six Sigma': ['black belt', 'capability analysis', 'control charts', 'doe', 'dmaic', 'fishbone',
                               'gage r&r', 'green belt', 'ishikawa', 'iso', 'kaizen', 'kpi', 'lean', 'metrics',
                               'pdsa', 'performance improvement', 'process improvement', 'quality',
                               'quality circles', 'quality tools', 'root cause', 'six sigma',
                               'stability analysis', 'statistical analysis', 'tqm'],
         'Operations management': ['automation', 'bottleneck', 'constraints', 'cycle time', 'efficiency', 'fmea',
                                   'machinery', 'maintenance', 'manufacture', 'line balancing', 'oee', 'operations',
                                   'operations research', 'optimization', 'overall equipment effectiveness',
                                   'pfmea', 'process', 'process mapping', 'production', 'resources', 'safety',
                                   'stoppage', 'value stream mapping', 'utilization'],
         'Supply chain': ['abc analysis', 'apics', 'customer', 'customs', 'delivery', 'distribution', 'eoq', 'epq',
                          'fleet', 'forecast', 'inventory', 'logistic', 'materials', 'outsourcing', 'procurement',
                          'reorder point', 'rout', 'safety stock', 'scheduling', 'shipping', 'stock', 'suppliers',
                          'third party logistics', 'transport', 'transportation', 'traffic', 'supply chain',
                          'vendor', 'warehouse', 'wip', 'work in progress'],
         'Project management': ['administration', 'agile', 'budget', 'cost', 'direction', 'feasibility analysis',
                                'finance', 'kanban', 'leader', 'leadership', 'management', 'milestones', 'planning',
                                'pmi', 'pmp', 'problem', 'project', 'risk', 'schedule', 'scrum', 'stakeholders'],
         'Data analytics': ['analytics', 'api', 'aws', 'big data', 'business intelligence', 'clustering', 'code',
                            'coding', 'data', 'database', 'data mining', 'data science', 'deep learning', 'hadoop',
                            'hypothesis test', 'iot', 'internet', 'machine learning', 'modeling', 'nosql', 'nlp',
                            'predictive', 'programming', 'python', 'r', 'sql', 'tableau', 'text mining',
                            'visualuzation'],
         'Healthcare': ['adverse events', 'care', 'clinic', 'cphq', 'ergonomics', 'healthcare',
                        'health care', 'health', 'hospital', 'human factors', 'medical', 'near misses',
                        'patient', 'reporting system']}

# Loop through PDF files in the folder and analyze each resume
for filename in os.listdir(resume_folder):
    if filename.endswith('.pdf'):
        resume_path = os.path.join(resume_folder, filename)
        resume_text = extract_text_from_pdf(resume_path)
        scores = analyze_resume_text(resume_text, terms)
        resume_scores.append([filename] + scores)

# Create a data frame with the scores summary for all resumes
columns = ['Filename'] + list(terms.keys())
resume_summary = pd.DataFrame(resume_scores, columns=columns)

# Print the summary for all resumes
print(resume_summary)

# Create pie chart visualization for a single resume (you can modify this for multiple resumes)
resume_scores = resume_summary.iloc[0, 1:]
pie = plt.figure(figsize=(10, 10))
plt.pie(resume_scores, labels=resume_scores.index, explode=(0.1, 0, 0, 0, 0, 0), autopct='%1.0f%%', shadow=True, startangle=90)
plt.title('Resume Decomposition by Areas')
plt.axis('equal')

# Display the pie chart in the Jupyter Notebook
plt.show()
```
